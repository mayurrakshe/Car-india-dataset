# Data Cleaning and Processing 

```python
# Fill missing numirical value  with the meadian 
numm_col = df.select_dtypes(include=["float64"]).columns
df[numm_col] = df[numm_col].fillna(df[numm_col].median())

```

```python
# Filling missing categriocal value with mode 

ctg_col = df.select_dtypes(include=["object"]).columns

df[ctg_col] = df[ctg_col].apply(lambda x: x.fillna(x.mode()[0]))
```

```python
# Verify that there are no missing value left 

missing_value_after = df .isnull().sum()

missing_value_after
```

## Predictive Modelingg 

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
# Prepare the Data 

x = df[numerical_features].drop(columns=["Rating"])
y = df["Rating"]

```

```python
# Split the Data into Tranning and Testing Sets 

X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Train a Random Forest Regressor

model = RandomForestRegressor(random_state = 42)
model.fit(X_train,y_train)
```

```python
# Make predictions 

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

# Clustering

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Standardize the Data 

scaler=StandardScaler()
X_scaled= scaler.fit_transform(x)
```

```python
# Perform K-mean clustring 

kmeans = KMeans(n_clusters=5,random_state=42)
kmeans.fit(X_scaled)
```

```python
# Add Cluster Labels to the Dataset
df["Cluster"] = kmeans.labels_
```

```python
# Visulize the Clusters

sns.scatterplot(x="Displacement",y="Fuel Efficiency",hue="Cluster",data=df,palette="viridis")
plt.title("Cluster of Cars")
plt.show()
```
